```python
import matplotlib.pyplot as plt
```

```python
with open('result.txt', 'r') as file:
    animal_amount = list(map(lambda x: (int(x[0]), int(x[1])), map(lambda x: x.split(", "), file.readlines()[1:])))

predators, prey = zip(*animal_amount)
plt.figure(figsize=(25, 10))
plt.plot(predators)
plt.plot(prey)
plt.ylabel('Population size', font={'size': 20})
plt.xlabel('Time', fontdict={'size': 20})
plt.legend(['Predators', 'Prey'], prop={'size': 15})
```
