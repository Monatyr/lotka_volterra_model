```python
import matplotlib.pyplot as plt
```

```python
with open('result.txt', 'r') as file:
    lines = file.readlines()
    animal_amount = list(map(lambda x: (int(x[0]), int(x[1])), map(lambda x: x.split(", "), lines[1:-2])))

predators, prey = zip(*animal_amount)
plt.figure(figsize=(15, 10))
plt.plot(predators)
plt.plot(prey)
plt.ylabel('Population size', font={'size': 20})
plt.xlabel('Time', fontdict={'size': 20})
plt.legend(['Predators', 'Prey'], prop={'size': 15})
```

```python
def plot_genotype(data, labels, title):
  a, b, c = data
  indices = list(range(len(a)))

  b = [b[i] + a[i] for i in range(len(b))]
  c = [c[i] + b[i] for i in range(len(c))]

  plt.figure(figsize=(15, 10))
  plt.fill_between(indices, c, alpha=0.8)
  plt.fill_between(indices, b, alpha=0.9)
  plt.fill_between(indices, a, alpha=0.75)
  plt.ylabel('Gene commoness', font={'size': 20})
  plt.xlabel('Time', fontdict={'size': 20})
  plt.legend(labels, prop={'size': 15})
  plt.title(title, fontdict={'size': 20})
  plt.show()
```

```python
prey_data = []
data = lines[-2].strip().split(":")

for i in range(3):
  non_empty = list(filter(None, data[i].split(",")))
  res = list(map(lambda x: float(x), non_empty))
  prey_data.append(res)


plot_genotype(prey_data, ['Reproduce', 'Eat', 'Run'], 'Prey genotype') #prey
```

```python
predator_data = []
data = lines[-1].strip().split(":")

for i in range(3):
  non_empty = list(filter(None, data[i].split(",")))
  res = list(map(lambda x: float(x), non_empty))
  predator_data.append(res)

plot_genotype(predator_data, ['Reproduce', 'Rest', 'Hunt'], "Predator genotype") #predators
```
